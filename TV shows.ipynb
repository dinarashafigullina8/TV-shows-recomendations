# Реализация рекомендаций

```python
recomendations=pd.DataFrame()
print(recomendations)
recList=list()
```

Реокмендация будет происходить по сервисам, на которых выходят фильмы и сериалы

```python
df = pd.pivot_table(df.drop(['Unnamed: 0', 'Age', 'IMDb', 'Rotten Tomatoes', 'Year', 'type'], axis=1),
               index=['Title'])
```

```python
df.head()
```

```python
for row in df:  
    k = 0 # Количество похожих объектов
    corrMatr=df.corrwith(df[row])
    corrMatr=pd.DataFrame(corrMatr)
    tempMatr=corrMatr
    tempMatr=tempMatr.drop([row],axis=0)
    while k != 1:
        name = tempMatr.idxmax().item()
        value = tempMatr[0][tempMatr.idxmax().item()]
        recList.append([row,name,value])
        tempMatr=tempMatr.drop([tempMatr.idxmax().item()],axis=0)
        k += 1
recomendations=recomendations.append(recList, ignore_index=True)
print(recomendations)
```

Для фильма\сериала, который выходит на Disney+ будет рекомендован фильм с этой же платформы и с Hulu.
Для фильма\сериала, который выходит на Netflix будет рекомендован фильм с этой же платформы и с Disney+.
Для фильма\сериала, который выходит на Prime Videoбудет рекомендован фильм с этой же платформы и с Disney+.
